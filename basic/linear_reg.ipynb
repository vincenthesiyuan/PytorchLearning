```python
%matplotlib inline
import torch
from IPython import display
from matplotlib import pyplot as plt
import numpy as np
import random

"""
loss function使用 square loss(平方损失)，即训练数据集中所有样本误差的平均来衡量模型预测质量
在模型训练中，希望找出一组模型参数，来使训练样本平均损失最小。

优化算法：
    analytical solution and numerical solution

    大多数深度学习模型并没有解析解，
    在数值解的优化算法中，mini-batch stochastic gradient descent 被广泛使用。

    hyperparameter:  learning rate, batch size

"""


"""
1. 生成数据集
"""
num_input = 2
num_examples = 1000
true_w = [2, -3.4]
true_b = 4.2
features = torch.randn(num_examples, num_input, dtype=torch.float32)
labels = true_w[0] * features[:, 0] + true_w[1] * features[:, 1] + true_b
labels += torch.tensor(np.random.normal(0, 0.01, size=labels.size()), dtype=torch.float32)
print(features[0], labels[0])

def use_svg_display():
    display.set_matplotlib_formats('svg')

def set_figsize(figsize = (3.5, 2.5)):
    use_svg_display()
    plt.rcParams['figure.figsize'] = figsize


set_figsize()
plt.scatter(features[:, 1].numpy(), labels.numpy(), 1);

#generate_dataset()
```

```python
"""
2. 读取数据
"""

def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices)
    for i in range(0, num_examples, batch_size):
        j = torch.LongTensor(indices[i: min(i + batch_size, num_examples)])
        yield features.index_select(0, j), labels.index_select(0, j)

batch_size = 10

for X, y in data_iter(batch_size, features, labels):
    print(X, y)
    break
```

```python
"""
3. 初始化模型参数
"""

w = torch.tensor(np.random.normal(0, 0.01, (num_input, 1))
                , dtype=torch.float32)
b = torch.zeros(1, dtype=torch.float32)

w.requires_grad_(requires_grad=True)
b.requires_grad_(requires_grad=True)
```

```python
"""
4. 定义模型
"""
def linreg(X, w, b):
    return torch.mm(X, w) + b
```

```python
"""
5. 定义损失函数
"""
def squared_loss(y_hat, y):
    return (y_hat - y.view(y_hat.size())) ** 2 / 2
```

```python
"""
6. 定义优化函数
"""
def sgd(params, lr, batch_size):
    for param in params:
        param.data -= lr * param.grad / batch_size
```

```python
"""
7. 训练模型
"""
lr = 0.03
num_epochs = 3
net = linreg
loss = squared_loss

for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y).sum()
        l.backward()  # batch的损失对模型参数求梯度
        sgd([w, b], lr, batch_size)  # 使用mini-batch随机梯度下降迭代模型
        
        w.grad.data.zero_()
        b.grad.data.zero_()
    
    train_l = loss(net(features, w, b), labels)
    # item(), 它可以将一个标量Tensor转换成一个Python number
    print('epoch %d, loss %f ' % (epoch + 1, train_l.mean().item()))
```
